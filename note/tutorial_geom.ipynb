```python
import symforce

symforce.set_backend("sympy")
symforce.set_log_level("warning")

from symforce.notebook_util import display
from symforce import sympy as sm
from symforce import geo
from symforce import ops

import numpy as np
```

```python
display(geo.Rot3())
```

```python
display(geo.Rot3.symbolic("R"))
```

```python
theta = sm.Symbol("theta")
R_mat = geo.Matrix(
    [
        [1, 0, 0],
        [0, sm.cos(theta), -sm.sin(theta)],
        [0, sm.sin(theta), sm.cos(theta)]
    ]
)
R = geo.Rot3.from_rotation_matrix(R_mat)
R_mat_recovered = R.to_rotation_matrix()


display(R_mat)
display(R)
display(R_mat_recovered) # R_matと同じになるはず
```

```python
display(R_mat.subs({"theta": 30.0 * np.pi / 180.0}))
display(R_mat_recovered.subs({"theta": 30.0 * np.pi / 180.0}))
```

```python
R = geo.Rot3.from_yaw_pitch_roll(0, 0, theta)
ypr = R.to_yaw_pitch_roll()

display(R)
display(ops.StorageOps.simplify(list(ypr))) # should be theta

theta_val = 30.0 / np.pi / 180.0
print(theta_val)
display(ops.StorageOps.simplify(list(ypr))[2].subs({"theta": theta_val}))
```

```python
R = geo.Rot3.from_angle_axis(angle=theta, axis=geo.Vector3(1, 0, 0))
display(R)
```

```python
world_R_body = geo.Rot3.symbolic("R")
body_t_point = geo.Vector3.symbolic("p")

world_t_point = world_R_body * body_t_point
display(world_R_body)
display(body_t_point)
display(world_t_point)
```

```python
body_R_cam = geo.Rot3.symbolic("R_cam") # this is identical to R_body_cam
world_R_cam = world_R_body * body_R_cam # this is identical to R_w_c = R_w_b * R_b_c

cam_R_body = body_R_cam.inverse()
display(body_R_cam)
display(cam_R_body)
```

```python
world_R_body_numeric = geo.Rot3.from_yaw_pitch_roll(0.1, -2.3, 0.7)
#             Rot3.from_angle_axis(yaw, V3(0, 0, 1))
#            * Rot3.from_angle_axis(pitch, V3(0, 1, 0))
#            * Rot3.from_angle_axis(roll, V3(1, 0, 0))
# this is identical to the intrinsic Yaw-Pitch-Roll(ZYX in scipy style notation) euler
display(world_t_point.subs(world_R_body, world_R_body_numeric))
```

```python
world_T_body = geo.Pose3.symbolic("T")
display(world_T_body)
```

```python
world_R_body = geo.Rot3.symbolic("R")
world_t_body = geo.Vector3.symbolic("t")
world_T_body = geo.Pose3(R=world_R_body, t=world_t_body)
display(world_T_body)
```

```python
body_T_cam = geo.Pose3.symbolic("T_cam")
world_T_cam = world_T_body * body_T_cam

body_t_point = geo.Vector3.symbolic("p")
world_t_point = world_T_body * body_t_point
display(world_t_point)
```

```python
body_T_world = world_T_body.inverse()
display(body_T_world)
display(world_T_body)
```
